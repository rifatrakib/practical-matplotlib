```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0.01, 5.0, 0.01)
s = np.exp(-t)
```

```python
fig, ax = plt.subplots()

ax.plot(t, s)
ax.set_xlim(5, 0) # decreasing time
ax.set_xlabel("decreasing time (s)")
ax.set_ylabel("voltage (mV)")
ax.set_title("Should be growing...")
ax.grid(True)

plt.show()
```
