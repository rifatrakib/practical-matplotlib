```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("a single plot")
plt.show()
```

```python
fig, axs = plt.subplots(2)
fig.suptitle("vertically stacked subplots")
axs[0].plot(x, y)
axs[1].plot(x, -y)
plt.show()
```
