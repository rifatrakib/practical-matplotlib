```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.set_xlim(300, 400)
ax.set_box_aspect(1)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)

ax1.plot([1, 5], [0, 10])
ax2.plot([100, 500], [10, 15])
ax1.set_box_aspect(1)
ax2.set_box_aspect(1)

plt.show()
```
