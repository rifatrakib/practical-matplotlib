```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.set_xlim(300, 400)
ax.set_box_aspect(1)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)

ax1.plot([1, 5], [0, 10])
ax2.plot([100, 500], [10, 15])
ax1.set_box_aspect(1)
ax2.set_box_aspect(1)

plt.show()
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot([0, 10])
ax2.plot([12, 10])
ax1.set_box_aspect(1)

plt.show()
```

```python
np.random.seed(19394790)
im = np.random.rand(16, 27)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, constrained_layout=True)

ax1.imshow(im)
ax2.plot([23, 45])
ax2.set_box_aspect(im.shape[0] / im.shape[1])

plt.show()
```

```python
x, y = np.random.rand(2, 400) * [[0.5], [180]]
gridspec_kw = {"height_ratios": [1, 3], "width_ratios": [3, 1]}
```

```python
fig, axs = plt.subplots(2, 2, sharex="col", sharey="row", gridspec_kw=gridspec_kw)

axs[0, 1].set_visible(False)
axs[0, 0].set_box_aspect(1 / 3)
axs[1, 0].set_box_aspect(1)
axs[1, 1].set_box_aspect(3 / 1)

axs[1, 0].scatter(x, y)
axs[0, 0].hist(x)
axs[1, 1].hist(y, orientation="horizontal")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.add_patch(plt.Circle((5, 3), 1))
ax.set_aspect("equal", adjustable="datalim")
ax.set_box_aspect(0.5)
ax.autoscale()

plt.show()
```

```python
fig, axs = plt.subplots(
    2, 3, subplot_kw={"box_aspect": 1},
    sharex=True, sharey=True, constrained_layout=True,
)

for i, ax in enumerate(axs.flatten()):
    ax.scatter(i % 3, -((i // 3) - 0.5) * 200, c=[plt.cm.hsv(i / 6)], s=300)

plt.show()
```
