```python
import warnings
import itertools
import matplotlib.pyplot as plt
```

```python
fontsizes = itertools.cycle([8, 16, 24, 32])
```

```python
def plotter(ax):
    ax.plot([1, 2])
    ax.set_xlabel("x-label", fontsize=next(fontsizes))
    ax.set_ylabel("y-label", fontsize=next(fontsizes))
    ax.set_title("title", fontsize=next(fontsizes))
```

```python
fig, ax = plt.subplots()
plotter(ax)
fig.tight_layout()
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
plotter(ax1)
plotter(ax2)
plotter(ax3)
plotter(ax4)
fig.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
plotter(ax1)
plotter(ax2)
fig.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
plotter(ax1)
plotter(ax2)
fig.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=3)
for ax in axs.flat:
    plotter(ax)
fig.tight_layout()
plt.show()
```
