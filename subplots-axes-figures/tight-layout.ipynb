```python
import warnings
import itertools
import matplotlib.pyplot as plt
```

```python
fontsizes = itertools.cycle([8, 16, 24, 32])
```

```python
def plotter(ax):
    ax.plot([1, 2])
    ax.set_xlabel("x-label", fontsize=next(fontsizes))
    ax.set_ylabel("y-label", fontsize=next(fontsizes))
    ax.set_title("title", fontsize=next(fontsizes))
```

```python
fig, ax = plt.subplots()
plotter(ax)
fig.tight_layout()
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
plotter(ax1)
plotter(ax2)
plotter(ax3)
plotter(ax4)
fig.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
plotter(ax1)
plotter(ax2)
fig.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
plotter(ax1)
plotter(ax2)
fig.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=3)
for ax in axs.flat:
    plotter(ax)
fig.tight_layout()
plt.show()
```

```python
plt.figure()
ax1 = plt.subplot(221)
ax2 = plt.subplot(223)
ax3 = plt.subplot(122)
plotter(ax1)
plotter(ax2)
plotter(ax3)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
ax1 = plt.subplot2grid((3, 3), (0, 0))
ax2 = plt.subplot2grid((3, 3), (0, 1), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 0), rowspan=2, colspan=2)
ax4 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
plotter(ax1)
plotter(ax2)
plotter(ax3)
plotter(ax4)
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()

gs1 = fig.add_gridspec(3, 1)
ax1 = fig.add_subplot(gs1[0])
ax2 = fig.add_subplot(gs1[1])
ax3 = fig.add_subplot(gs1[2])
plotter(ax1)
plotter(ax2)
plotter(ax3)
gs1.tight_layout(fig, rect=[None, None, 0.45, None])

gs2 = fig.add_gridspec(2, 1)
ax4 = fig.add_subplot(gs2[0])
ax5 = fig.add_subplot(gs2[1])
plotter(ax4)
plotter(ax5)

with warnings.catch_warnings():
    # gs2.tight_layout cannot handle the subplots from the first gridspec (gs1),
    # so it will raise a warning
    # we are going to match the gridspec manually so we can filter the warning away
    warnings.simplefilter("ignore", UserWarning)
    gs2.tight_layout(fig, rect=[0.45, None, None, None])

# now match the top and bottom of two gridspecs
top = min(gs1.top, gs2.top)
bottom = max(gs1.bottom, gs2.bottom)

gs1.update(top=top, bottom=bottom)
gs2.update(top=top, bottom=bottom)

plt.show()
```
