```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2 * np.pi * t)
s2 = np.sin(4 * np.pi * t)
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(t, s1)
plt.subplot(212)
plt.plot(t, 2 * s1)
plt.show()
```

```python
plt.figure(2)
plt.plot(t, s2)
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(t, s2, "s")
ax = plt.gca()
ax.set_xticklabels([])
plt.show()
```
