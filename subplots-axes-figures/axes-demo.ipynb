```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# fixing random state for reproducibility
np.random.seed(19872400)
```

```python
# create some data to use for the plot
dt = 0.001
t = np.arange(0.0, 1.0, dt)
r = np.exp(-t[:1000] / 0.05) # impulse response
x = np.random.randn(len(t))
s = np.convolve(x, r)[:len(x)] * dt # colored noise
```

```python
fig, ax = plt.subplots()

ax.plot(t, s)
ax.set_xlim(0, 1)
ax.set_ylim(1.1 * np.min(s), 2 * np.max(s))
ax.set_xlabel("time (s)")
ax.set_ylabel("current (nA)")
ax.set_title("Gaussian colored noise")

# this is an inset axes over the main axes
ax1 = fig.add_axes([0.65, 0.6, 0.2, 0.2], facecolor="k")
ax1.hist(s, 400, density=True)
ax1.set(title="probability", xticks=[], yticks=[])

# this is another inset axes over the main axes
ax2 = fig.add_axes([0.2, 0.6, 0.2, 0.2], facecolor="k")
ax2.plot(t[:len(r)], r)
ax.set(title="impulse response", xlim=(0, 0.2), xticks=[], yticks=[])

plt.show()
```
