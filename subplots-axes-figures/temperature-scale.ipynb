```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fahrenheit_to_celcius(temp):
    """returns temperature in Celsius given Fahrenheit temperature"""
    return (5.0 / 9.0) * (temp - 32)
```

```python
def plotter():
    # define a closure function to register as a callback
    def convert_ax_c_to_celcius(ax_f):
        """update second axis according with first axis"""
        y1, y2 = ax_f.get_ylim()
        ax_c.set_ylim(fahrenheit_to_celcius(y1), fahrenheit_to_celcius(y2))
        ax_c.figure.canvas.draw()
    
    fig, ax_f = plt.subplots()
    ax_c = ax_f.twinx()
    
    # automatically update ylim of ax2 when ylim of ax1 changes
    ax_f.callbacks.connect("ylim_changed", convert_ax_c_to_celcius)
    ax_f.plot(np.linspace(-40, 120, 100))
    ax_f.set_xlim(0, 100)
    
    ax_f.set_title("two scales: fahrenheit and celcius")
    ax_f.set_ylabel("fahrenheit")
    ax_f.set_ylabel("celcius")
    
    plt.show()
```

```python
plotter()
```
