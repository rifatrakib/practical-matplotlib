```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = np.meshgrid(np.arange(0, 2 * np.pi, 0.2), np.arange(0, 2 * np.pi, 0.2))
u = np.cos(x)
v = np.sin(y)
```

```python
fig, ax = plt.subplots()
ax.set_title("arrows scale with plot width, not view")
q = ax.quiver(x, y, u, v, units="width")
qk = ax.quiverkey(q, 0.9, 0.9, 2, r"$2 \frac{m}{s}$", labelpos="E", coordinates="figure")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_title("pivot='mid'; every third arrow; units='inches'")
q = ax.quiver(x[::3, ::3], y[::3, ::3], u[::3, ::3], v[::3, ::3], pivot="mid", units="inches")
qk = ax.quiverkey(q, 0.9, 0.9, 1, r"$1 \frac{m}{s}$", labelpos="E", coordinates="figure")
ax.scatter(x[::3, ::3], y[::3, ::3], color="r", s=5)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_title("pivot='tip'; scales with x view")
m = np.hypot(u, v)
q = ax.quiver(x, y, u, v, m, units="x", pivot="tip", width=0.022, scale=1 / 0.15)
qk = ax.quiverkey(q, 0.9, 0.9, 1, r"$1 \frac{m}{s}$", labelpos="E", coordinates="figure")
ax.scatter(x, y, color="0.5", s=1)
plt.show()
```
