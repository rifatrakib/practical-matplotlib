```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def formatter(pct, all_values):
    absolute = int(np.round(pct / 100 * np.sum(all_values)))
    return f"{pct:.1f}%\n({absolute} g)"
```

```python
recipe = ["375 g flour", "75 g sugar", "250 g butter", "300 g berries"]
data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]
```

```python
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw={"aspect": "equal"})

wedges, text, autotexts = ax.pie(
    data, autopct=lambda pct: formatter(pct, data), textprops={"color": "w"},
)
ax.legend(
    wedges, ingredients, title="ingredients",
    loc="center left", bbox_to_anchor=(1, 0, 0.5, 1),
)
plt.setp(autotexts, size=8, weight="bold")
ax.set_title("pie bakery")

plt.show()
```
