```python
import matplotlib.pyplot as plt
```

```python
labels = ["frogs", "hogs", "dogs", "logs"]
fracs = [15, 20, 45, 10]
```

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].pie(fracs, labels=labels, autopct="%1.1f%%", shadow=True)

axs[0, 1].pie(fracs, labels=labels, autopct="%.0f%%", shadow=True, explode=(0, 0.1, 0, 0))

patches, texts, autotexts = axs[1, 0].pie(
    fracs, labels=labels, autopct="%.0f%%", radius=0.5,
    textprops={"size": "smaller"}, shadow=True,
)
# make percent texts even smaller
plt.setp(autotexts, size="x-small")
autotexts[0].set_color("white")

patches, texts, autotexts = axs[1, 1].pie(
    fracs, labels=labels, autopct="%.0f%%", shadow=False, radius=0.5,
    explode=(0, 0.05, 0, 0), textprops={"size": "smaller"},
)
plt.setp(autotexts, size="x-small")
autotexts[0].set_color("white")

plt.show()
```
