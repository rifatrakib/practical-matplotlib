```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.linspace(0, 3, 301)
theta = 2 * np.pi * r
angle = np.deg2rad(67.5)
bbox_to_anchor = (0.5 + np.cos(angle) / 2, 0.5 + np.sin(angle) / 2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="polar", facecolor="lightgoldenrodyellow")

ax.plot(theta, r, color="tab:orange", lw=3, label="filled line")
ax.plot(0.5 * theta, r, color="tab:blue", ls="--", lw=3, label="dotted line")
ax.tick_params(grid_color="palegoldenrod")
ax.legend(loc="lower left", bbox_to_anchor=bbox_to_anchor)

plt.show()
```
