```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(19278399)
```

```python
n = 150
r = 2 * np.random.rand(n)
theta = 2 * np.pi * np.random.rand(n)
area = 200 * r ** 2
colors = theta
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="polar")
ax.scatter(theta, r, c=colors, s=area, cmap="hsv", alpha=0.75)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="polar")

ax.scatter(theta, r, c=colors, s=area, cmap="hsv", alpha=0.75)
ax.set_rorigin(-2.5)
ax.set_theta_zero_location("W", offset=10)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="polar")

ax.scatter(theta, r, c=colors, s=area, cmap="hsv", alpha=0.75)
ax.set_thetamin(45)
ax.set_thetamax(135)

plt.show()
```
