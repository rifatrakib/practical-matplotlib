```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(199237890)
```

```python
n_bins = 20
n_points = 387
labels = ["A", "B", "C"]

data = [
    np.random.normal(0, 1, n_points),
    np.random.normal(6, 1, n_points),
    np.random.normal(-3, 1, n_points),
]

hist_range = (np.min(data), np.max(data))
binned_data = [np.histogram(d, range=hist_range, bins=n_bins)[0] for d in data]
binned_maximums = np.max(binned_data, axis=1)
x_locations = np.arange(0, sum(binned_maximums), np.max(binned_maximums))

bin_edges = np.linspace(hist_range[0], hist_range[1], n_bins + 1)
heights = np.diff(bin_edges)
centers = bin_edges[:-1] + heights / 2
```

```python
fig, ax = plt.subplots()

for x_loc, binned_set in zip(x_locations, binned_data):
    lefts = x_loc - 0.5 * binned_set
    ax.barh(centers, binned_set, height=heights, left=lefts)

ax.set_xticks(x_locations, labels)
ax.set_xlabel("data sets")
ax.set_ylabel("vales")

plt.show()
```
