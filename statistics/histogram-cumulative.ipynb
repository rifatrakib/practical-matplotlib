```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(198732090)
```

```python
mu = 200
sigma = 25
n_bins = 50
x = np.random.normal(mu, sigma, size=100)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

n, bins, patches = ax.hist(
    x, n_bins, density=True, histtype="step", cumulative=True, label="empirical"
)

y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu)) ** 2))
y = y.cumsum()
y /= y[-1]

ax.plot(bins, y, "k--", linewidth=1.5, label="theoritical")

ax.hist(x, bins=bins, density=True, histtype="step", cumulative=-1, label="reversed empirical")
ax.grid(True)
ax.legend(loc="right")
ax.set_title("cumulative step histograms")
ax.set_xlabel("annual rainfall (mm)")
ax.set_ylabel("likelihood of occurance")

plt.show()
```
