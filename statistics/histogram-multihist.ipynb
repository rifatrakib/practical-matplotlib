```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1972910)
```

```python
n_bins = 10
x = np.random.randn(1000, 3)
colors = ["red", "tan", "lime"]
x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)

ax1.hist(x, n_bins, density=True, histtype="bar", color=colors, label=colors)
ax1.legend(prop={"size": 10})
ax1.set_title("bars with legend")

ax2.hist(x, n_bins, density=True, histtype="bar", stacked=True)
ax2.set_title("stacked bar")

ax3.hist(x, n_bins, histtype="step", stacked=True, fill=False)
ax3.set_title("stack step (unfilled)")

ax4.hist(x_multi, n_bins, histtype="bar")
ax4.set_title("different sample sizes")

fig.tight_layout()
plt.show()
```
