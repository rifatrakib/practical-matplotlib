```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1974920)
```

```python
n = 100000
x = np.random.standard_normal(n)
y = 2.0 + 3.0 * x + 4.0 * np.random.standard_normal(n)
xlim = x.min(), x.max()
ylim = y.min(), y.max()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True, figsize=(9, 4))

hb = ax1.hexbin(x, y, gridsize=50, cmap="inferno")
ax1.set(xlim=xlim, ylim=ylim)
ax1.set_title("hexagon binning")
cb = fig.colorbar(hb, ax=ax1, label="counts")

hb = ax2.hexbin(x, y, gridsize=50, bins="log", cmap="inferno")
ax2.set(xlim=xlim, ylim=ylim)
ax2.set_title("with a log color scale")
cb = fig.colorbar(hb, ax=ax2, label="log10(N)")

plt.show()
```
