```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import EventCollection
```

```python
# fixing random state for reproducibility
np.random.seed(19675801)
```

```python
# create random data
data = np.random.random([2, 10])
```

```python
# split the data into two parts
xdata1 = data[0, :]
xdata2 = data[1, :]

# sort the data so it makes clean curves
xdata1.sort()
xdata2.sort()
```

```python
# create some y data points
ydata1 = xdata1 ** 2
ydata2 = 1 - xdata2 ** 3
```

```python
# plot the data
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.plot(xdata1, ydata1, color="tab:blue")
ax.plot(xdata2, ydata2, color="tab:orange")

# create the events marking the x data points
xevents_1 = EventCollection(xdata1, color="tab:blue", linelength=0.05)
xevents_2 = EventCollection(xdata2, color="tab:orange", linelength=0.05)

# create the events marking the y data points
yevents_1 = EventCollection(ydata1, color="tab:blue", linelength=0.05, orientation="vertical")
yevents_2 = EventCollection(ydata2, color="tab:orange", linelength=0.05, orientation="vertical")

# add the events to the axis
ax.add_collection(xevents_1)
ax.add_collection(xevents_2)
ax.add_collection(yevents_1)
ax.add_collection(yevents_2)

# set the limits
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])

ax.set_title("line plot with data points")
plt.show()
```
