```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import EngFormatter
```

```python
prng = np.random.RandomState(12993470)
```

```python
xs = np.logspace(1, 9, 100)
ys = (0.8 + 0.4 * prng.uniform(size=100)) * np.log10(xs) ** 2
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(7, 10))

for ax in (ax1, ax2):
    ax.set_xscale("log")

ax1.set_title("full unit ticklabels, w/ default precision & space separator")
formatter1 = EngFormatter(unit="Hz")
ax1.xaxis.set_major_formatter(formatter1)
ax1.plot(xs, ys)
ax1.set_xlabel("frequency")

ax2.set_title("SI-prefix only ticklabels, 1-digit precision & thin space separator")
formatter2 = EngFormatter(places=1, sep="\N{THIN SPACE}") # U+2009
ax2.xaxis.set_major_formatter(formatter2)
ax2.plot(xs, ys)
ax2.set_xlabel("frequency (Hz)")

plt.tight_layout()
plt.show()
```
