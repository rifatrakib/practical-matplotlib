```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l1 = np.array((1, 1))
l2 = np.array((5, 5))
angle = 45
```

```python
fig, ax = plt.subplots()

h = ax.plot(range(0, 10), range(0, 10))
ax.set_xlim([-10, 20])

th1 = ax.text(
    *l1, "text not rotated correctly", fontsize=16,
    rotation=angle, rotation_mode="anchor",
)
th2 = ax.text(
    *l2, "text rotated correctly", fontsize=16, rotation=angle,
    rotation_mode="anchor", transform_rotates_text=True,
)

plt.show()
```
