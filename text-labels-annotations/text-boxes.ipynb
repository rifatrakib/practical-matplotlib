```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(19749290)
```

```python
x = 30 * np.random.randn(10000)
mu = x.mean()
median = np.median(x)
sigma = x.std()
textstr = "\n".join((
    r"$\mu=%.2f$" % (mu, ),
    r"$\mathrm{median}=%.2f$" % (median, ),
    r"$\sigma=%.2f$" % (sigma, ),
))
props = {"boxstyle": "round", "facecolor": "wheat", "alpha": 0.5}
```

```python
fig, ax = plt.subplots()
ax.hist(x, 50)
ax.text(
    0.05, 0.95, textstr, transform=ax.transAxes,
    fontsize=14, va="top", bbox=props,
)
plt.show()
```
