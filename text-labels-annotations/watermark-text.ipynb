```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(19348920)
```

```python
fig, ax = plt.subplots()
ax.plot(np.random.rand(20), "-o", ms=20, lw=2, alpha=0.7, mfc="orange")
ax.grid()
ax.text(
    0.5, 0.5, "created with matplotlib", transform=ax.transAxes, fontsize=40,
    color="gray", alpha=0.5, ha="center", va="center", rotation=30,
)
plt.show()
```
