```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# fixing random state for reproducibility
np.random.seed(1979450)
```

```python
n = 100
r0 = 0.6
x = 0.9 * np.random.rand(n)
y = 0.9 * np.random.rand(n)

# 0 to 10 point radii
area = (20 * np.random.rand(n)) ** 2
c = np.sqrt(area)
r = np.sqrt(x ** 2 + y ** 2)

area_1 = np.ma.masked_where(r < r0, area)
area_2 = np.ma.masked_where(r >= r0, area)

theta = np.arange(0, np.pi / 2, 0.01)
```

```python
plt.scatter(x, y, s=area_1, marker="^", c=c)
plt.scatter(x, y, s=area_2, marker="o", c=c)

# show the boundary between the regions:
plt.plot(r0 * np.cos(theta), r0 * np.sin(theta))
plt.show()
```
