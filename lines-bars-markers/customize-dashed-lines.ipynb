```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 500)
y = np.sin(x)
```

```python
fig, ax = plt.subplots()

# using set_dashes() to modify dashing of an existing line
line_1, = ax.plot(x, y, label="using set_dashes()")
line_1.set_dashes([2, 2, 10, 2]) # 2pt line, 2pt break, 10pt line, 2pt break

# using plot(..., dashes=...) to set the dashing when creating a line
line_2, = ax.plot(x, y - 0.2, dashes=[6, 2], label="using the dashes parameter")

ax.legend()
plt.show()
```
