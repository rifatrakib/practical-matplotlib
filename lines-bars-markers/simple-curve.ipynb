```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)
```

```python
fig, ax = plt.subplots()

ax.plot(t, s)
ax.set(xlabel="time (s)", ylabel="voltage (mV)", title="time vs voltage")
ax.grid()

plt.show()
```
