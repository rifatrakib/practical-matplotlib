```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# using fill_betweenx to color along the horizontal direction between two curves
y = np.arange(0.0, 2, 0.01)
x1 = np.sin(2 * np.pi * y)
x2 = 1.2 * np.sin(4* np.pi * y)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(6, 6))

ax1.fill_betweenx(y, 0, x1)
ax1.set_title("between (x1, 0)")

ax2.fill_betweenx(y, x1, 1)
ax2.set_title("between (x1, 1)")
ax2.set_xlabel("x")

ax3.fill_betweenx(y, x1, x2)
ax3.set_title("between (x1, x2)")

plt.show()
```

```python
# now fill between x1 and x2 where a logical condition is met
x3 = np.ma.masked_greater(x2, 1.0)
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, sharey=True, figsize=(6, 6))

ax1.plot(x1, y, x2, y, color="black")
ax1.fill_betweenx(y, x1, x2, where=x2 >= x1, facecolor="green")
ax1.fill_betweenx(y, x1, x2, where=x2 < x1, facecolor="red")
ax1.set_title("fill_betweenx where")

# test support for masked arrays
ax2.plot(x1, y, x3, y, color="black")
ax2.fill_betweenx(y, x1, x3, where=x3 >= x1, facecolor="green")
ax2.fill_betweenx(y, x1, x3, where=x3 < x1, facecolor="red")
ax2.set_title("regions with x2 > 1 are masked")
```
