```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define a list of markevery cases to plot
cases = [
    None, 8, (30, 8), [16, 24, 32], [0, -1],
    slice(100, 200, 3), 0.1, 0.4, (0.2, 0.4),
]

# data points
delta = 0.11
x = np.linspace(0, 10 - 2 * delta, 200) + delta
y = np.sin(x) + 1.0 + delta
```

```python
# markevery with linear scales
fig, axs = plt.subplots(3, 3, figsize=(10, 6), constrained_layout=True)

for ax, markevery in zip(axs.flat, cases):
    ax.set_title(f"{markevery=}")
    ax.plot(x, y, "o", ls="-", ms=4, markevery=markevery)

plt.show()
```

```python
# markevery with log scales
fig, axs = plt.subplots(3, 3,figsize=(10, 6), constrained_layout=True)

for ax, markevery in zip(axs.flat, cases):
    ax.set_title(f"{markevery=}")
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.plot(x, y, "o", ls="-", ms=4, markevery=markevery)

plt.show()
```

```python
# markevery on zoomed plots
fig, axs = plt.subplots(3, 3, figsize=(10, 6), constrained_layout=True)

for ax, markevery in zip(axs.flat, cases):
    ax.set_title(f"{markevery=}")
    ax.plot(x, y, "o", ls="-", ms=4, markevery=markevery)
    ax.set_xlim((6, 6.7))
    ax.set_ylim((1.1, 1.7))

plt.show()
```
