```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.1, 2 * np.pi, 41)
y = np.exp(np.sin(x))
```

```python
plt.stem(x, y)
plt.show()
```

```python
markerline, stemlines, baseline = plt.stem(x, y, linefmt="grey", markerfmt="D", bottom=1.1)
markerline.set_markerfacecolor("none")
plt.show()
```
